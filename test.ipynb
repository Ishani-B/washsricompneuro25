```python
import numpy as np 
import matplotlib.pyplot as plt

time = np.linspace(0, 6, 601)
input_signal = np.sin(2 * np.pi * 1 * time)

threshold = 0.5 


output_spikes = []
for val in input_signal:
    if val > threshold:
        output_spikes.append(1)
    else:
        output_spikes.append(0)
output_spikes = np.array(output_spikes)

plt.figure(figsize=(8,4))
plt.plot(time, input_signal, label='Input signal')
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold = {threshold}')
plt.step(time, output_spikes, where='mid', label='Neuron firing (output)')
plt.xlabel('Time (s)')
plt.ylabel('Signal / Spike')
plt.legend()
plt.show()
```
